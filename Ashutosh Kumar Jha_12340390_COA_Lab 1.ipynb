```python
def binaryToDecimal(binary,n):
        if binary[0] == '1':
            return int(binary, 2) - (1 << n)
        else:
            return int(binary, 2)
```

```python
def boothMultiplication(n, binary1, binary2):
    M = binaryToDecimal(binary1, n)
    Q = binaryToDecimal(binary2, n)

    A = 0
    Q_1 = 0
    count = n

    while count > 0:
        if (Q & 1) == 1 and Q_1 == 0:
            A = A - M
        elif (Q & 1) == 0 and Q_1 == 1:
            A = A + M

        A, Q, Q_1 = (A >> 1), (Q >> 1) | ((A & 1) << (n - 1)), (Q & 1)


        count -= 1

    result = (A << n) | Q
    result_binary = format(result & ((1 << (2 * n)) - 1), f'0{2 * n}b')

    return result_binary
```

```python
def main():
  n = int(input("Enter the number of bits you want to use: "))
  binary1 = input("Enter the first binary number: ")
  binary2 = input("Enter the second binary number: ")
  result = boothMultiplication(n,binary1,binary2)
  print(f"Result in binary: {result}")
  print(f"Result in decimal: {binaryToDecimal(result,n)}")
```
